```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
initial_ammount = np.array([10000000,20000000,50000000,0])
```

```python
def dispatch(initial_ammount,amount):
    max_amount=np.max(initial_ammount)
    work_amount = max_amount+1 - initial_ammount
    tot = np.sum(work_amount)
    dispatch_values = np.asarray(np.round((amount*work_amount)/tot), dtype=int)
    initial_ammount += dispatch_values
    return initial_ammount
```

```python
dispatch(initial_ammount, 200)
```

```python
run_number = 400
initial_ammount = np.array([10000,2000,50000,0])
amounts = [[],[],[],[]]
for j in range(4):
    amounts[j].append(initial_ammount[j])

value = 500

for i in range(run_number):
    initial_ammount = dispatch(initial_ammount, value)
    for j in range(4):
        amounts[j].append(initial_ammount[j])
```

```python

plt.plot( range(run_number+1),amounts[0])
plt.plot( range(run_number+1),amounts[1])
plt.plot( range(run_number+1),amounts[2])
plt.plot( range(run_number+1),amounts[3])
```
